```python
import matplotlib.pyplot as plt
import numpy as np

with open("./dataset/eil51.txt") as f:
    cities = f.read().strip().split("\n")

cities = [tuple(map(int, city.strip().split(" ")[1:])) for city in cities]

def get_2city_dist(c1, c2):
    return np.sqrt((c1[0] - c2[0])**2 + (c1[1] - c2[1])**2)

```

```python
with open("path.txt") as f:
    paths = f.read().strip().split("\n")

for path in paths:
    pathlist = path.strip().split(" ")
    if len(set(pathlist)) != 51:
        print("error")
        
```

```python
with open("path.txt") as f:
    best_paths = [list(map(int, path.strip().split(" "))) for path in f.read().strip().split("\n")]

for best_path in best_paths:
    for i in range(len(best_path)):
        c1, c2 = best_path[i], best_path[(i+1) % len(best_path)]
        c1 = cities[c1]
        c2 = cities[c2]
        plt.plot([c1[0], c2[0]], [c1[1], c2[1]], color="red")
    plt.show()
```
