```python
import matplotlib.pyplot as plt
import numpy as np

with open("./dataset/eil51.txt") as f:
    cities = f.read().strip().split("\n")

cities = [tuple(map(int, city.strip().split(" ")[1:])) for city in cities]

def get_2city_dist(c1, c2):
    return np.sqrt((c1[0] - c2[0])**2 + (c1[1] - c2[1])**2)

```
