```python
import matplotlib.pyplot as plt
```

```python
dims = (2, 10, 30)
test_funcs = ("ackley", "griewank", "rastrigin", "schwefel", "sphere", "sum_squares", "zakharov", "rosenbrock", "michalewicz", "powell")
```

```python
for test_func in test_funcs:
    plt.figure(figsize=(30, 6))
    for i, dim in enumerate(dims):
        with open(f"./{test_func}/de_{test_func}_{dim}d_51r.txt") as f:
            de_data = f.read()

        de_data = de_data[:de_data.index("\n\n")]
        de_data = list(map(eval, de_data.strip().split("\n")))

        with open(f"./{test_func}/lshade_{test_func}_{dim}d_51r.txt") as f:
            lshade_data = f.read()

        lshade_data = lshade_data[:lshade_data.index("\n\n")]
        lshade_data = list(map(eval, lshade_data.strip().split("\n")))

        plt.subplot(int(f"13{i+1}"))
        plt.title(f"{test_func.capitalize()} - {dim}D")
        plt.plot(de_data, label="DE")
        plt.plot(lshade_data, label="L-SHADE")
        plt.xlabel("Evaluation")
        plt.ylabel("Objective value")
        plt.legend()

    plt.savefig(f"./_imgs/{test_func}.jpg", dpi=150)
    plt.show()
```
